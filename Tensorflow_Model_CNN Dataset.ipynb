```python
# %% [markdown]
# # CNN Report: 8 Conv Layers + 5 Pooling Layers (Colab-runnable)
#
# **Author:** Auto-generated
#
# **Summary:** This notebook builds an 8-convolutional-layer + 5-pooling-layer CNN in TensorFlow/Keras, trains briefly on CIFAR-10, and provides step-by-step visualizations:
# - model summary
# - training curves
# - convolutional filters (weights)
# - activation maps after each conv/pool
# - Grad-CAM saliency map
#
# Paste this file into Google Colab and run cells. For a quick run, set `EPOCHS = 2` or reduce dataset size.

# %%
# Step 1 — Environment check & imports
# Explanation: Import TensorFlow, plotting and helper libs. Check TF and GPU availability.

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import os
import math
from typing import List

print("TensorFlow version:", tf.__version__)
print("GPU available:", bool(tf.config.list_physical_devices('GPU')))

# %%
# Step 2 — Load and preprocess dataset (CIFAR-10)
# Explanation: Load CIFAR-10, normalize to [0,1], and create a validation split.

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# normalize
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32')  / 255.0

# quick validation split for demos
VAL_SIZE = 5000
x_val = x_train[-VAL_SIZE:]
y_val = y_train[-VAL_SIZE:]
x_train = x_train[:-VAL_SIZE]
y_train = y_train[:-VAL_SIZE]

print('train:', x_train.shape, 'val:', x_val.shape, 'test:', x_test.shape)

# make tf.data pipelines
BATCH_SIZE = 64
train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(10000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((x_val, y_val)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# %%
# Step 3 — Build model: 8 conv layers, 5 pooling layers (Functional API)
# Explanation: Names conv1..conv8 and pool1..pool5 to simplify visualization.

inputs = layers.Input(shape=(32, 32, 3))
# conv1 -> pool1
x = layers.Conv2D(32, 3, activation='relu', padding='same', name='conv1')(inputs)
x = layers.MaxPooling2D(2, name='pool1')(x)
# conv2, conv3 -> pool2
x = layers.Conv2D(32, 3, activation='relu', padding='same', name='conv2')(x)
x = layers.Conv2D(64, 3, activation='relu', padding='same', name='conv3')(x)
x = layers.MaxPooling2D(2, name='pool2')(x)
# conv4, conv5 -> pool3
x = layers.Conv2D(64, 3, activation='relu', padding='same', name='conv4')(x)
x = layers.Conv2D(128, 3, activation='relu', padding='same', name='conv5')(x)
x = layers.MaxPooling2D(2, name='pool3')(x)
# conv6, conv7 -> pool4
x = layers.Conv2D(128, 3, activation='relu', padding='same', name='conv6')(x)
x = layers.Conv2D(256, 3, activation='relu', padding='same', name='conv7')(x)
x = layers.MaxPooling2D(2, name='pool4')(x)
# conv8 -> pool5
x = layers.Conv2D(256, 3, activation='relu', padding='same', name='conv8')(x)
x = layers.MaxPooling2D(2, name='pool5')(x)
# head
x = layers.GlobalAveragePooling2D(name='gap')(x)
x = layers.Dense(128, activation='relu', name='dense1')(x)
outputs = layers.Dense(10, activation='softmax', name='predictions')(x)

model = models.Model(inputs=inputs, outputs=outputs, name='cnn_8conv_5pool')

# %%
# Step 4 — Compile and inspect model
# Explanation: Compile with Adam and sparse categorical crossentropy, then show the summary.

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# %%
# Step 5 — Train the model (brief demo)
# Explanation: Use callbacks. Lower EPOCHS for fast Colab runs.

EPOCHS = 6  # set to 2-3 for quicker demo
callbacks = [
    tf.keras.callbacks.ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy'),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1)
]

history = model.fit(train_ds, epochs=EPOCHS, validation_data=val_ds, callbacks=callbacks)

# %%
# Step 6 — Plot training curves (loss & accuracy)
# Explanation: Quick visualization to confirm training behavior.

def plot_history(history):
    fig, axes = plt.subplots(1,2, figsize=(12,4))
    axes[0].plot(history.history['loss'], label='train loss')
    axes[0].plot(history.history['val_loss'], label='val loss')
    axes[0].set_title('Loss')
    axes[0].legend()

    axes[1].plot(history.history['accuracy'], label='train acc')
    axes[1].plot(history.history['val_accuracy'], label='val acc')
    axes[1].set_title('Accuracy')
    axes[1].legend()
    plt.show()

plot_history(history)

# %%
# Step 7 — List conv/pool layers (for later visualization)
# Explanation: Gather layer names in model order so visualization code can iterate in that order.

layer_names = [layer.name for layer in model.layers if (('conv' in layer.name) or ('pool' in layer.name))]
print('conv/pool layers in model order:')
for i, n in enumerate(layer_names, 1):
    print(f"{i:02d}", n)

# %%
# Step 8 — Visualize convolutional filters (weights)
# Explanation: Show the first few filters of any conv layer. RGB kernels shown for first layer.

def show_conv_weights(layer_name: str, max_filters=6):
    layer = model.get_layer(layer_name)
    W = layer.get_weights()[0]  # (kh, kw, in_ch, out_ch)
    kh, kw, in_ch, out_ch = W.shape
    n = min(out_ch, max_filters)
    plt.figure(figsize=(n*2, 2))
    for i in range(n):
        filt = W[:, :, :, i]
        if in_ch == 3:
            img = (filt - filt.min()) / (filt.max() - filt.min() + 1e-9)
        else:
            img = np.mean(filt, axis=-1)
            img = (img - img.min()) / (img.max() - img.min() + 1e-9)
        ax = plt.subplot(1, n, i+1)
        ax.imshow(img.squeeze())
        ax.axis('off')
        ax.set_title(f'{layer_name}\\nfilter {i}')
    plt.show()

# Example: show filters from conv1 and conv3
show_conv_weights('conv1', max_filters=6)
show_conv_weights('conv3', max_filters=6)

# %%
# Step 9 — Build activation model to obtain intermediate outputs
# Explanation: Create a Keras model that returns activations for all conv/pool layers.

from tensorflow.keras.models import Model
outputs_for_vis = [model.get_layer(name).output for name in layer_names]
activation_model = Model(inputs=model.input, outputs=outputs_for_vis)

# choose a sample image from test set
sample_idx = 0
sample_img = x_test[sample_idx:sample_idx+1]

activations = activation_model.predict(sample_img)
print('Activations computed for', len(activations), 'layers')

# %%
# Step 10 — Visualize activation maps after each conv/pool for the sample image
# Explanation: For each activation tensor show a few feature maps and the mean map.

def show_activation_maps(activations, layer_names, maps_per_layer=6):
    for act, name in zip(activations, layer_names):
        h, w, c = act.shape[1], act.shape[2], act.shape[3]
        n_maps = min(maps_per_layer, c)
        fig_cols = n_maps + 1
        plt.figure(figsize=(2*fig_cols, 2))
        for i in range(n_maps):
            fmap = act[0, :, :, i]
            ax = plt.subplot(1, fig_cols, i+1)
            ax.imshow(fmap, interpolation='nearest')
            ax.axis('off')
            ax.set_title(f'{name}\\nmap {i}')
        mean_map = np.mean(act[0], axis=-1)
        ax = plt.subplot(1, fig_cols, fig_cols)
        ax.imshow(mean_map, interpolation='nearest')
        ax.axis('off')
        ax.set_title(f'{name}\\nmean')
        plt.show()

# Run the visualization (reduce maps_per_layer to limit plots)
show_activation_maps(activations, layer_names, maps_per_layer=6)

# %%
# Step 11 — Grad-CAM for class explanation
# Explanation: Compute Grad-CAM heatmap for the final conv layer (conv8) and overlay on input.

def make_gradcam_heatmap(img_tensor, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_tensor)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-9)
    heatmap = tf.image.resize(heatmap[..., tf.newaxis], (img_tensor.shape[1], img_tensor.shape[2]))
    heatmap = tf.squeeze(heatmap).numpy()
    return heatmap

def overlay_heatmap_on_image(img, heatmap, alpha=0.4):
    import matplotlib.cm as cm
    cmap = cm.get_cmap('jet')
    heatmap_rgba = cmap(heatmap)
    heatmap_rgb = heatmap_rgba[..., :3]
    overlay = img * (1 - alpha) + heatmap_rgb * alpha
    overlay = np.clip(overlay, 0, 1)
    return overlay

# pick test image and compute heatmap
sample_img = x_test[sample_idx:sample_idx+1]
preds = model.predict(sample_img)
pred_class = np.argmax(preds[0])
print('Predicted class:', pred_class)
heatmap = make_gradcam_heatmap(sample_img, model, last_conv_layer_name='conv8', pred_index=pred_class)

plt.figure(figsize=(8,4))
plt.subplot(1,3,1); plt.title('Input image'); plt.imshow(sample_img[0]); plt.axis('off')
plt.subplot(1,3,2); plt.title('Heatmap'); plt.imshow(heatmap); plt.axis('off')
plt.subplot(1,3,3); plt.title('Overlay'); plt.imshow(overlay_heatmap_on_image(sample_img[0], heatmap)); plt.axis('off')
plt.show()

# %%
# Step 12 — Save final model and notes
model.save('cnn_8conv_5pool_final.h5')
print('Saved model to cnn_8conv_5pool_final.h5')

print('\nNotes:')
print(' - To use your own dataset, replace Step 2 with image_dataset_from_directory or a custom tf.data pipeline.')
print(' - To speed things up, reduce EPOCHS or use fewer training samples while debugging.')

# End of notebook
```
